```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
# * Total Schools
# * Total Students
# * Total Budget
# * Average Math Score
# * Average Reading Score
# * % Passing Math (The percentage of students that passed math.)
# * % Passing Reading (The percentage of students that passed reading.)
# * % Overall Passing (The percentage of students that passed math **and** reading.)

    
school_data_complete



```

```python
total_schools = len(school_data_complete["school_name"].unique())
total_students = school_data_complete["student_name"].count()
total_budget = school_data_complete["budget"].sum()
avg_math_score = school_data_complete["math_score"].mean()
avg_reading_score = school_data_complete["reading_score"].mean()
# math passing calculation

math_pass_counts = len(school_data_complete[school_data_complete["math_score"] >= 70])
perc_math = (math_pass_counts * 100)/total_students

# Reading percentage calculation
read_pass_counts = len(school_data_complete[school_data_complete["reading_score"] >= 70])
perc_read = (read_pass_counts * 100)/total_students

# Overall passing counts & percentage


overall_counts = len(school_data_complete[(school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70)])

perc_overall = (overall_counts * 100)/total_students


#Total Schools	Total Students	Total Budget	Average Math Score	Average Reading Score
#% Passing Math	% Passing Reading	% Overall Passing

dist_summary = {"Total Schools" : [total_schools],
                "Total Student" : [total_students],
                "Total Budget" : total_budget,
                "Average Math Score" : avg_math_score,
                "Average Reading Score" : avg_reading_score,
                "% Passing Math" : perc_math,
                "% Passing Reading" : perc_read,
                "% Overall Passing" : perc_overall
                }

pd.DataFrame(dist_summary)

```
